```python
import numpy as np
import matplotlib.pyplot as plt
import os
os.chdir(r'D:\lym本科\科研相关\华为项目资料\2025-1-23\Encoder_PELT')
```

```python
def generate_arma(mu, gamma, theta, noise_cov, length, p, q):
    y = np.zeros(length)
    epsilon = np.random.multivariate_normal(np.zeros(len(mu)), noise_cov, size=length)
    
    for t in range(max(p, q), length):
        y[t] = mu + np.dot(gamma, y[t-p:t][::-1]) + np.dot(theta, epsilon[t-q:t][::-1]) + epsilon[t]
    
    return y
```

```python
total_length = 500  # 总长度
segment_length = 150  # 每段长度
np.random.seed(2025)

# 初始化时间序列
time_series = []
changepoints = [150,300,450]

# 初始参数
mu = np.array([0.0])  # 均值
gamma = np.array([0.5,0.2])  # 自回归系数
theta = np.array([0.2,0.2])  # 移动平均系数
noise_cov = np.array([[1.0]])  # 噪声协方差矩阵
p = 2
q = 2

# 生成时间序列
for i in range(0, total_length, segment_length):
    # 生成一段数据
    segment = generate_arma(mu, gamma, theta, noise_cov, segment_length, p, q)
    time_series.extend(segment)
    
    # 动态改变模型参数
    mu += np.random.normal(1, 1, size=mu.shape)  # 随机改变mu
    # mu += 1
    print(f"Segment {i//segment_length + 1}: mu={mu}, gamma={gamma}, theta={theta}")

# 绘制生成的时间序列
plt.figure(figsize=(10, 6))
plt.plot(time_series, label='Generated Time Series')
plt.title('Generated Multivariate Time Series with Dynamic ARMA Parameters')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()

np.save(f'./data/ARMA/ARMA_series_500.npy',time_series)
np.save(f'./data/ARMA/ARMA_series_500_changepoints.npy',changepoints)
```
