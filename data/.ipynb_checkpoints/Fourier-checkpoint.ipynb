```python
def create_windows(data, window_size):
    windows = []
    for i in range(len(data) - window_size + 1):
        window = data[i:i+window_size,:]
        windows.append(window)
    return np.array(windows)
```

```python
def add_anomalies(data, num_anomalies=5, anomaly_strength=10):
    anomalous_data = data.copy()
    num_samples = data.shape[1]

    anomaly_mask = np.zeros(num_samples, dtype=int)
    anomaly_positions = np.random.choice(
        num_samples, 
        num_anomalies, 
        replace=False
    ).tolist()

    anomaly_mask[anomaly_positions] = 1
    
    for pos in anomaly_positions:
        for i in range(data.shape[0]):
            direction = np.random.choice([-1, 1])
            anomalous_data[i, pos] += direction * anomaly_strength

    return anomalous_data, anomaly_mask
```

```python
np.random.seed(2026)
num_samples = 200
num_frequencies = 10
max_frequency = 50 
var_num = 5
noise_std = 0 

frequencies = np.random.uniform(1, max_frequency, (var_num, num_frequencies))
amplitudes = np.random.uniform(1, 20, (var_num, num_frequencies))
phases = np.random.uniform(0, 2 * np.pi, (var_num, num_frequencies))

t = np.linspace(0, 1, num_samples)
data = np.zeros((var_num, num_samples))
print("data:",data.shape)


for i in range(var_num):
    for freq, amp, phase in zip(frequencies[i], amplitudes[i], phases[i]):
        data[i] += amp * np.sin(2 * np.pi * freq * t + phase)
    noise = np.random.normal(0, noise_std, num_samples)
    data[i] += noise
    
plt.figure(figsize=(12, 8))
# for i in range(var_num):
#     plt.plot(t, data[0], label=f'Time Series {1}')
plt.plot(t, data[0])
# plt.title('Data Generated from Fourier Basis')
# plt.xlabel('Time')
# plt.ylabel('Amplitude')
# plt.legend()
plt.grid(True)
plt.show()

window_size = 10
data = data.T
data_window = create_windows(data, window_size)
print("time_series_window:",data_window.shape)
data_window = data_window.reshape(data_window.shape[0],data_window.shape[2]*window_size)
test_unchanged_data = data_window[150:,]
data_window =  np.delete(data_window, np.arange(150, 191), axis=0)
print("time_series_window:",data_window.shape)

np.save(f'./data/Fourier/Fourier_series_200_train_v4.npy', data_window)
```

```python
np.random.seed(2042)
num_samples = 200
num_frequencies = 10
max_frequency = 150
var_num = 5 
noise_std = 0 

changepoints = [40, 82, 124] 

frequencies = np.random.uniform(100, max_frequency, (var_num, num_frequencies))
amplitudes = np.random.uniform(1, 20, (var_num, num_frequencies))
phases = np.random.uniform(0, 2 * np.pi, (var_num, num_frequencies))

t = np.linspace(0, 1, num_samples)
data = np.zeros((var_num, num_samples)) 

for i in range(var_num):
    for j in range(num_samples):
        if j in changepoints:
            if j >= changepoints[0]:
                frequencies[i] = np.random.uniform(200, 250, num_frequencies)
                amplitudes[i] = np.random.uniform(1, 20, num_frequencies)
                phases[i] = np.random.uniform(0, 2 * np.pi, num_frequencies)
            if j >= changepoints[1]:
                frequencies[i] = np.random.uniform(150, 200, num_frequencies)
                amplitudes[i] = np.random.uniform(1, 20, num_frequencies)
                phases[i] = np.random.uniform(0, 2 * np.pi, num_frequencies)
            if j >= changepoints[2]:
                frequencies[i] = np.random.uniform(300, 350, num_frequencies)
                amplitudes[i] = np.random.uniform(1, 20, num_frequencies)
                phases[i] = np.random.uniform(0, 2 * np.pi, num_frequencies)

        signal = 0
        for freq, amp, phase in zip(frequencies[i], amplitudes[i], phases[i]):
            signal += amp * np.sin(2 * np.pi * freq * t[j] + phase)

        noise = np.random.normal(0, noise_std)
        data[i, j] = signal + noise

data_with_anomalies, anomaly_list = add_anomalies(data, num_anomalies=20, anomaly_strength=10)
print("data_with_anomalies:",data_with_anomalies.shape)
print("anomaly_list:",anomaly_list)

plt.figure(figsize=(12, 8))
# for i in range(var_num):
#     plt.plot(t, data_with_anomalies[i], label=f'Time Series {i + 1}')
plt.plot(t, data_with_anomalies[0])
plt.plot(t, data_with_anomalies[0], '--', alpha=0.5, label='Time Series with Anomalies')
anomaly_positions = np.where(anomaly_list == 1)[0]
plt.scatter(t[anomaly_positions], data_with_anomalies[0, anomaly_positions], 
            color='red', s=100, label='Anomalies', zorder=5)
plt.title('Data Generated from Fourier Basis with Changepoints')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

window_size = 10
data_with_anomalies = data_with_anomalies.T
data_window = create_windows(data_with_anomalies, window_size)
print("time_series_window:",data_window.shape)
data_window = data_window.reshape(data_window.shape[0],data_window.shape[2]*window_size)
data_window = np.concatenate((test_unchanged_data, data_window), axis=0)
new_arr = anomaly_list[:num_samples - window_size + 1]
print("new_arr:",len(new_arr))
anomaly_list = np.pad(new_arr, (test_unchanged_data.shape[0], 0), 'constant')
print("anomaly_list:",anomaly_list)
print("anomaly_list:",len(anomaly_list))
print("time_series_window:",data_window.shape)

np.save(f'./data/Fourier/Fourier_series_200_test_v4.npy', data_window)
np.save(f'./data/Fourier/Fourier_series_200_test_label_v4.npy', anomaly_list)
```
